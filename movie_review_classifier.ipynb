```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("moviereviews.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
len(df)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#The above doesnt check for blank strings. It only checks for NaN
```

```python
len(df[df['review'].str.isspace()])
#This will tell us how many blank spaces are there
#We will need to drop them also
```

```python
#dropping the rows with reviews as blank space
df=df[~df['review'].str.isspace()]
```

```python
#check
len(df[df['review'].str.isspace()])
```

```python
df['label'].value_counts()
```

```python
df.head()
```

```python
X=df['review']
y=df['label']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.3 , random_state=101)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf.fit(X_train)
```

```python
X_train_tfidf = tfidf.transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
model = LinearSVC(dual=True)
```

```python
model.fit(X_train_tfidf , y_train)
```

```python
preds = model.predict(X_test_tfidf)
```

```python
print(classification_report(preds , y_test) )
```
